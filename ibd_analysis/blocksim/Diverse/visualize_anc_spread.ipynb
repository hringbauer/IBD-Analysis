# Script for Visualization of block sharing.
In this script, the idea behind block sharing is visualized.
This is done in Jupyter, so that one can flexibly explore different options.

Idea for Visualization: Show grid of chromosomes (with x- and y-Vector); and color code the spread of Ancestry.
First for 1D; second for 2D

```python
# Do the Imports:
import matplotlib.pyplot as plt
import sys
import numpy as np

sys.path.append('../')                      # Add Parent Directory (with Grid...)
sys.path.append('../../analysis_popres/')   # Add the Analyse Popres Directory (2x up) 
from grid import factory_Grid
```

```python
def plot_ancestry(grid_x=10, grid_y=10, time=1, figsize=(12,12), 
                  block_list=[], total_length=1, two_blocks=False):
    '''Function to plot where Ancestry resides.
    Chromosomes sit from 0 to grid_x; and 0 to grid_y.
    Total Length: The Length of an IBD block.
    Two Blocks: Whether 2 blocks have to be plot'''
    # First try to plot the background grid (in grey). (DELETE THIS LATER ON)
    
    
    # First define some constant
    offset = 0.1 # Offset on the y-Axis
    dist1 = 0.16 # Around the x-Axis
    color_bg = "silver"   # Color for the lines
    lw = 14 # How broad the lines are
    bl_c="Red"  # Color of the Block
    bl_c1, bl_c2 = "Red", "Blue"  # Colors of Blocks in case of 2 of them
    fs=24 # Fontsize
    
    plt.figure(figsize=figsize)
    #plt.plot(figsize=figsize)
    for x in xrange(grid_x):
        for y in xrange(grid_y):
            plt.plot([x-dist1, x-dist1],[y+offset,y+1-offset], 
                     color=color_bg, linewidth=lw, solid_capstyle="butt")
            plt.plot([x+dist1, x+dist1],[y+offset,y+1-offset], 
                     color=color_bg, linewidth=lw, solid_capstyle="butt")
    
    # Set some Plot Properties
    ax=plt.gca()
    ax.set_aspect(1) # Set the aspect ratio to fixed value:
    plt.xticks([])
    plt.yticks([])
    
    margin=0.6 # Margin around the pic
    plt.xlim([-margin-offset, grid_x-1+margin+dist1])
    plt.ylim([-margin,grid_y+margin-offset])
        
    # Some Code for coordinate arrows:
    hw=0.1 #
    hl=0.3 # Length of the arrow head
    ap=-0.2
    al=2 # Length of the arrow
    plt.arrow(ap-dist1-0.1, ap, 0, al, head_width=hw, head_length=hl, fc='k', ec='k')
    plt.arrow(ap-dist1-0.1, ap, al, 0, head_width=hw, head_length=hl, fc='k', ec='k')
    plt.annotate("x", (ap-dist1+al+1.5*hl, ap), fontsize=fs)
    plt.annotate("y", (ap-dist1-0.2, ap+al+1.5*hl), fontsize=fs)
    
    # Plot the time
    plt.annotate(r"$t=%i$" %time, (grid_x-2, ap-0.1), fontsize=fs)
    
    def plot_line_segment(x_pos, y_pos, ch, start, end, total_length=1, color="Red"):
        '''Quick helper function to plot a short line segment
        pos: Where in space; start: Where the block starts, length: How long it is
        ch: Which chromosome (0 or 1), total_length: Scale Factor to rescale block sharing'''
        assert(0<=start<total_length)
        assert(start<end<=total_length)
        
        
        # Do the Renormalization:
        total_length = float(total_length)
        scale_factor = 1-(2 * offset) # Scale Factor - plotted chromosomes are shorter
        start = (start / total_length) * scale_factor
        end = (end / total_length) * scale_factor
        
        if ch==0:
            ch=-1
            
        # Actually plot the Block:
        plt.plot([x_pos+ch*dist1, x_pos+ch*dist1], [y_pos+offset+start, y_pos+offset+end], 
                 linewidth=lw, color=color, solid_capstyle="butt", alpha=0.8)
    
    # Extract the unique origins:
    origins = tuple(set([block[5] for block in block_list]))
    print("Blocks of %i different Origins detected." % len(origins))
    print(origins)
    
    # Plot the Blocks:
    for block in block_list:
        x, y, ch = block[0], block[1], block[2]
        start, end = block[3], block[4]
        origin = block[5]  
        
        # Give the block the right color.
        color = bl_c
        if two_blocks == True:
            assert(len(origins)>=2) # Sanity Check
            if origin == origins[0]:
                color = bl_c1  
            elif origin == origins[1]:
                color = bl_c2
            else:
                print("Too many origins!")

        plot_line_segment(x, y, ch, start, end, total_length, color)
    
    plt.show()
```

## Run the Simulation for one Chromosome
Trace back Ancestry for one chromosome. Plot its spread of Ancestry

```python
# Run the Simulation.
# Which times to take:
t_vec = np.array([0,1,2,3,5,10])

gw = 11

#1) Set up the Grid
grid = factory_Grid("classic")

# Set the grid Parameters:
grid.dispmode = "normal"  # Set the Dispersal Mode
grid.delete = False
grid.set_gridwidth(grid_width=gw)
grid.set_sigma(sigma=1.0)
grid.chrom_l=150.0
grid.rec_rate=100.0
grid.reset_grid()  # Delete everything and initializes Grid

# Set the Sample
grid.set_samples([(5,5,0),])

# Update the Grid:
        



# 2) Run and plot
t_diff = t_vec[1:]-t_vec[:-1]

# Runs the Simulations
for delta_t in t_diff:
    bl=grid.give_anc_blocks()  # Extract Ancestry
    t=grid.t                   # Member the time!
    plot_ancestry(gw, gw, time=t, block_list=bl, total_length=150)
    
    grid.update_t(delta_t)     # Do the actual run

# The last step
bl=grid.give_anc_blocks()  # Extract Ancestry
t=grid.t                   # Member the time!
plot_ancestry(gw, gw, time=t, block_list=bl, total_length=150)
```

## Run the Simulation for 2 Chromosomes
Trace back the Ancestry of two chromosomes; and visualize where they overlap

```python
# Run the Simulation.
# Which times to take:
t_vec = np.array([0,1,2,3,5,10])

gw = 11

#1) Set up the Grid
grid = factory_Grid("classic")

# Set the grid Parameters:
grid.dispmode = "normal"  # Set the Dispersal Mode
grid.delete = False
grid.set_gridwidth(grid_width=gw)
grid.set_sigma(sigma=1.0)
grid.chrom_l=150.0
grid.rec_rate=100.0
grid.reset_grid()  # Delete everything and initializes Grid
grid.IBD_treshold = 0.0

# Set the Sample
grid.set_samples([(4,5,0),])   # Set Sample #1
grid.set_samples([(6,5,0),])   # Set Sample #2

# 2) Run and plot
t_diff = t_vec[1:]-t_vec[:-1]

# Runs the Simulations
for delta_t in t_diff:
    bl=grid.give_anc_blocks()  # Extract Ancestry
    t=grid.t                   # Member the time!
    plot_ancestry(gw, gw, time=t, block_list=bl, total_length=150, two_blocks=True)
    grid.update_t(delta_t)     # Do the actual run

# The last step
bl=grid.give_anc_blocks()  # Extract Ancestry
t=grid.t                   # Member the time!
plot_ancestry(gw, gw, time=t, block_list=bl, total_length=150, two_blocks=True)
```
